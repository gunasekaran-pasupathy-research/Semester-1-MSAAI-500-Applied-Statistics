```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.__version__
```

```python
students = pd.read_table('http://stat4ds.rwth-aachen.de/data/Students.dat',
                          sep="\\s+")
```

```python
students.head()
```

```python
gender = students["gender"].replace({0: "male", 1: "female"})
```

```python
students.rename(columns={"gender": "a", "B": "c"})
```

```python
gender
```

```python
gender.value_counts(1).round(2).plot(kind="bar", width=0.9, rot=0)
plt.title("Student Gender Proportion Comparison")
plt.xlabel("Gender")
plt.ylabel("Proportion")
plt.show()

gender.value_counts()
```
