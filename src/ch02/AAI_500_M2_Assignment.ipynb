$$

**(a) Show that the joint probability distribution of diagnosis and disease status is as shown in
Table 2.6. Given that a diagnostic test result is positive, explain how this joint distribution
shows that the 12% of incorrect diagnoses for the 99% of women not having breast cancer
swamp the 86% of correct diagnoses for the 1% of women actually having breast cancer.**

Using the definition of conditional probability, P(+ and D) = P(+ | D) · P(D), and similarly for the other three combinations:

P(+ and D) = 0.86 * 0.01 = 0.0086
P(− and D) = 0.14 * 0.01 = 0.0014
P(+ and Dc) = 0.12 * 0.99 = 0.1188
P(− and Dc) = 0.88 * 0.99 = 0.8712

Given a positive test, the positive predictive value is
P(D | +) = P(+ and D) / P(+) = 0.0086 / (0.0086 + 0.1188) = 0.0086 / 0.1274 = 0.0675....

So only about 6.75% of positive results truly have cancer. The 12% false-positive rate acts on the 99% of women without cancer, producing 0.1188, which is far more than the 0.0086 from the 86% sensitivity acting on the 1% who actually have cancer.

**(b) The first test for detecting HIV-positive status had a sensitivity of 0.999 and specificity
of 0.9999. Explain what these mean. If at that time 1 in 10,000 men were truly HIV-positive,
find the positive predictive value. Based on this example, explain the potential disadvantage
of routine diagnostic screening of a population for a rare disease.**
Sensitivity 0.999 = 99.9% of HIV-positive men test positive. Specificity 0.9999 = 99.99% of HIV-negative men test negative.

With P(HIV) = 1/10000:
P(+) = (0.999)(0.0001) + (0.0001)(0.9999) = 0.0002
P(HIV | +) = (0.999 × 0.0001) / 0.0002 = 0.50
So only about 50% of positive results truly correspond to HIV. Disadvantage: for a rare disease, even a near-perfect test produces roughly as many false positives as true positives, so most positives are false alarms that cause unnecessary anxiety, follow-up testing, and cost when applied to the general population.


```python
# ---------- Part (a): Mammogram ----------

prevalence_of_cancer = 0.01  # 1% of women have breast cancer
prob_no_cancer = 1 - prevalence_of_cancer  # 99%

sensitivity = 0.86  # P(positive | has cancer)
false_positive_rate = 0.12  # P(positive | no cancer)
specificity = 1 - false_positive_rate  # 0.88

# Joint probabilities = conditional × marginal  (Table 2.6)
prob_cancer_and_positive_test = sensitivity * prevalence_of_cancer  # 0.0086
prob_cancer_and_negative_test = (1 - sensitivity) * prevalence_of_cancer  # 0.0014
prob_no_cancer_and_positive_test = false_positive_rate * prob_no_cancer  # 0.1188
prob_no_cancer_and_negative_test = specificity * prob_no_cancer  # 0.8712

print("Joint distribution (Table 2.6):")
print(f"P(cancer,positive) = {prob_cancer_and_positive_test:.4f}")
print(f"P(cancer,negative) = {prob_cancer_and_negative_test:.4f}")
print(f"P(no cancer, positive) = {prob_no_cancer_and_positive_test:.4f}")
print(f"P(no cancer, negative) = {prob_no_cancer_and_negative_test:.4f}")

# Marginal probability of a positive test
prob_positive_test = ( prob_cancer_and_positive_test + prob_no_cancer_and_positive_test)

# Bayes' Rule: P(cancer | positive test)
prob_cancer_given_positive_test = (
        prob_cancer_and_positive_test / prob_positive_test
)

print(f"\nP(positive test)            = {prob_positive_test:.4f}")
print(f"P(cancer | positive test)   = {prob_cancer_given_positive_test:.4f} "
      f"(~{prob_cancer_given_positive_test * 100:.2f}%)")

# ---------- Part (b): HIV screening ----------

prevalence_of_hiv = 1 / 10_000  # 1 in 10,000 men are HIV-positive
prob_not_hiv = 1 - prevalence_of_hiv

hiv_test_sensitivity = 0.999  # P(positive | HIV-positive)
hiv_test_specificity = 0.9999  # P(negative | HIV-negative)
hiv_test_false_positive_rate = 1 - hiv_test_specificity

prob_hiv_test_positive = (
        hiv_test_sensitivity * prevalence_of_hiv
        + hiv_test_false_positive_rate * prob_not_hiv
)

positive_predictive_value = (
        (hiv_test_sensitivity * prevalence_of_hiv) / prob_hiv_test_positive
)

print("\nHIV screening:")
print(f"  P(test positive)                 = {prob_hiv_test_positive:.6f}")
print(f"  PPV = P(HIV | test positive)     = {positive_predictive_value:.4f} "
      f"(~{positive_predictive_value * 100:.2f}%)")
```

## Bayes’ Theorem and AI Decision Reliability

An autonomous drone uses an onboard AI to detect whether an area contains a **target object** (for example, a missing person wearing a bright jacket).  
The system raises an alert when it believes a target is present. However, the model is not perfect.

- $P(\text{Target}) = 0.03$
- $P(\text{Alert} \mid \text{Target}) = 0.95$
- $P(\text{Alert} \mid \text{No Target}) = 0.10$

### Tasks

1. **Compute the probability of an alert**

   Use the law of total probability:
   $$P(\text{Alert}) = P(\text{Alert} \mid \text{Target})P(\text{Target}) + P(\text{Alert} \mid \text{No Target})P(\text{No Target})$$

2. **Compute the posterior probability**

   Apply Bayes’ theorem.

$$P(\text{Target} \mid \text{Alert}) =
\frac{P(\text{Alert} \mid \text{Target})P(\text{Target})}{P(\text{Alert})}$$



3. **Implement in Python**

   Write a short script that calculates and prints both probabilities with clear labels.

4. **Interpret your result**

   In a short paragraph (as a code comment), explain what your computed $(P(\text{Target} \mid \text{Alert})$ means about the AI's reliability when it triggers an alert.

5. **Bonus**

   Create a function that allows a user to change:
   - $P(\text{Target})$
   - $P(\text{Alert} \mid \text{Target})$
   - $P(\text{Alert} \mid \text{No Target})$

   Then return the updated posterior probability and show how the result changes when $P(\text{Target}) = 0.005$.

### Starter Code

```python
## The code seems to be already working, so, I tried to interpret the results
# P(Alert): 0.1255
# P(Target | Alert): 0.2271
# Posterior with P(Target)=0.005: 0.04556354916067146

# P(Alert): 0.1255 & P(Target | Alert): 0.2271  means that when the drone raises an alert, there is only about a 23% chance that a real target is actually present. In other words, roughly 77% of alerts are false alarms. Even though the model is 95% sensitive and the false-positive rate is only 10%, targets are rare (3% base rate), so the 10% false alarms generated from the 97% empty areas vastly outnumber the true detections from the 3% target areas. This is the same "rare-event swamping" effect seen in the mammogram and HIV examples — the positive predictive value depends heavily on the prior.

## Bayes' Theorem in AI Detection Reliability

############################# Given probabilities ##############################
P_target = 0.03
P_alert_given_target = 0.95
P_alert_given_no_target = 0.10
################################################################################
# Compute complements
P_no_target = 1 - P_target

# 1. Total probability of alert
P_alert = (P_alert_given_target * P_target) + (P_alert_given_no_target * P_no_target)

# 2. Posterior probability using Bayes' theorem
P_target_given_alert = (P_alert_given_target * P_target) / P_alert

print(f"P(Alert): {P_alert:.4f}")
print(f"P(Target | Alert): {P_target_given_alert:.4f}")


# Interpretation:
# When the AI raises an alert, this tells us how likely it is that
# a target is actually present. A low posterior means many false alarms.

# Bonus: parameterized function
def bayes_posterior(P_target, P_alert_given_target, P_alert_given_no_target):
    P_no_target = 1 - P_target
    P_alert = (P_alert_given_target * P_target) + (P_alert_given_no_target * P_no_target)
    return (P_alert_given_target * P_target) / P_alert


print("Posterior with P(Target)=0.005:",
      bayes_posterior(0.005, 0.95, 0.10))


```
